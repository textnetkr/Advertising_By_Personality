# KeyBERT

```python
from keybert import KeyBERT
```

```python
doc = """짝짝짝! 친구 맺으면 1,OOOP♥	정답만 맞히면 포인트 주는 이벤트!\\퀴즈 맞히고 1,OOOP 바로 받아가기☞"""
```

```python
kw_model = KeyBERT()
keywords = kw_model.extract_keywords(doc)
```

```python
print(keywords)
```
