# Alpaca Few-Shot Test

```python
import torch
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM
)

device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
model_name = "beomi/KoAlpaca-Polyglot"
```

```python
# tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# model
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
from pshmodule.processing import processing as p
```

```python
nlp = p.Nlp()
```

```python
j_unicode = nlp.convert_to_other_unicode(text)
print(f"json unicode : {j_unicode}")
```

predict

```python
encoded = tokenizer(j_unicode)

sample = {k: torch.tensor([v]).to(device) for k, v in encoded.items()}
del sample['token_type_ids']
print(sample)
```

```python
with torch.no_grad():
    pred = model.generate(
        **sample,
        penalty_alpha=0.6,
        top_k=5,
        max_length=512,
        eos_token_id=tokenizer.eos_token_id,
    )
```

```python
print(tokenizer.decode(pred[0], skip_special_tokens=True))
```
