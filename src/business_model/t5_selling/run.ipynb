```python
import json
import pandas as pd
```

```python
data = []

with open('drive/MyDrive/Advertising_By_Personality/data/train.json', 'r', encoding='utf-8') as f:
    for line in f:
        data.append(json.loads(line.rstrip('\n|\r')))
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
sum([len(i) for i in df.label.tolist()]) / len(df)
```

```python
from transformers import T5TokenizerFast, T5ForConditionalGeneration
```

```python
tokenizer = T5TokenizerFast.from_pretrained("paust/pko-t5-base")
```

```python
df.head()
```

input

```python
temp = [f"Text Generation: <type>{i[1]['type']}<classified>{i[1]['classified']}<advertiser>{i[1]['advertiser']}<sender>{i[1]['sender']}<product>{i[1]['product']}<product_detail>{i[1]['product_detail']}<purpose>{i[1]['purpose']}<benefit>{i[1]['benefit']}<period>{i[1]['period']}<target>{i[1]['target']}<season>{i[1]['season']}<weather>{i[1]['weather']}<anniv>{i[1]['anniv']}<selling_point>{i[1]['selling_point']}" for i in df.iterrows()]
```

```python
input = tokenizer(
    temp,
    max_length=512,
    padding="max_length",
    truncation=True,
    return_tensors="np",
)
```

```python
result = [len([j for j in i if j !=0]) for i in input['input_ids'].tolist()]
```

```python
sum(result) / len(result)
```

output

```python
output = tokenizer(
    [t + tokenizer.eos_token for t in df.label.tolist()],
    max_length=512,
    padding="max_length",
    truncation=True,
    return_tensors="np",
)
```

```python
result = [len([j for j in i if j !=0]) for i in output['input_ids'].tolist()]
```

```python
sum(result) / len(result)
```
